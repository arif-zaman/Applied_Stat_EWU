```r
library(ggplot2)
library(dendextend)
library(dplyr)
library(cluster)
library(purrr)
```

```r
mt_tbl = tbl_df(mtcars)
str(mt_tbl)
```

# Scaling Dataset
        each column has its own unit. so we need to scale the dataset for cluster analysis.

```r
df = scale(mt_tbl)
head(df)
```

# Hierarchical Clustering

```r
dist_df = dist(df)

# Using "complete" linkage :: other options - "single" or "average"
hc_dist_df = hclust(dist_df, "complete")
plot(hc_dist_df)
```

## Dendrogram based on height (h)

```r
dend_df = as.dendrogram(hc_dist_df)
dend_colored = color_branches(dend_df, h=6)
plot(dend_colored)
```

## Dendrogram based on number of Cluster (k)

```r
dend_df = as.dendrogram(hc_dist_df)
dend_colored = color_branches(dend_df, k=3)
plot(dend_colored)
```

### From the dendrogram, it seems 2 cluster is better fit.

```r
cluster_assignments = cutree(hc_dist_df, k=2) ## or h=7 .. (height)
mt_tbl_cluster = mt_tbl %>% mutate(cluster=cluster_assignments)
```

## Summary Statistics of the Clusters

```r
count(mt_tbl_cluster, cluster)
```

### Median Value within the clusters

```r
mt_tbl_cluster %>% group_by(cluster) %>% summarise_all(funs(median(.)))
```

# K-Means & PAM Clustering

## Elbow Plot

```r
total_withinss <- map_dbl(1:15,  function(k){
  model <- kmeans(x = mt_tbl, centers = k)
  model$tot.withinss
})

# Generate a data frame containing both k and tot_withinss
elbow_df <- data.frame(
  k = 1:15,
  total_withinss = total_withinss
)

# Plot the elbow plot
ggplot(elbow_df, aes(x = k, y = total_withinss)) +
  geom_line() +
  scale_x_continuous(breaks = 1:15)
```

## Silhouette Plot (PAM)

```r
sil_width <- map_dbl(2:15,  function(k){
  model <- pam(mt_tbl, k = k)
  model$silinfo$avg.width
})


sil_df <- data.frame(
  k = 2:15,
  sil_width = sil_width
)

ggplot(sil_df, aes(x = k, y = sil_width)) +
  geom_line() +
  scale_x_continuous(breaks = 2:15)
```

```r
pam_model_2 <- pam(mt_tbl, k = 2)
pam_model_10 <- pam(mt_tbl, k = 10)
pam_model_12 <- pam(mt_tbl, k = 12)

par(mfrow=c(1,3))
plot(silhouette(pam_model_2))
plot(silhouette(pam_model_10))
plot(silhouette(pam_model_12))
```

#### Both Silhouette and Elbow plot suggest cluster size of 2.

## Kmeans Cluster with cluster size=2

```r
model = kmeans(mt_tbl, centers = 2)
mt_tbl_kmeans = mt_tbl %>% mutate(cluster=model$cluster) 
```

## Summary Statistics

```r
count(mt_tbl_kmeans, cluster)
```

### Median Value within the clusters

```r
mt_tbl_kmeans %>% group_by(cluster) %>% summarise_all(funs(median(.)))
```
